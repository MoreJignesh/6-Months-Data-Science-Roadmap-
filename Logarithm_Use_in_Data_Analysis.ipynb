```python
#we need to import the library
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
```

```python
table_data = [['Company','Revenue'],
      ['Tesla','31.0'],
      ['Uber','11.0'],
      ['Amazon','386.0'],
      ['RBC','11.5'],
      ['TD','11.9'],
      ['BMO','10.7'],
      ['Scotia','11.2']]

for row in table_data:
  for col in table_data:
    print(col, end='\t')
  print()
```

```python
df = (tabulate(table_data,headers="firstrow",tablefmt="fancy_grit"))
print(df)
```

```python
#converting data into pandas dataframe
df = pd.DataFrame(table_data[1:], columns=table_data[0])

# Convert 'Revenue' column to float
df['Revenue'] = df['Revenue'].astype(float)

# Plot the DataFrame
df.plot(x='Company', y='Revenue', kind='bar')
```

```python
# As we have notice the Amazon revenue is quite high so we are not able to focus on rest other as it seems no difference between them.
# So we have to use Log to better understand the difference between them

log10 = df.plot(x='Company', y='Revenue', kind='bar', logy= True)
```
